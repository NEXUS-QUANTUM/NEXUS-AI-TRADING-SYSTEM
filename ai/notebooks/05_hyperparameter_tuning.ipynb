```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pickle
import os
import warnings
warnings.filterwarnings('ignore')

# Scikit-learn
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Modèles
try:
    import xgboost as xgb
    XGB_AVAILABLE = True
except ImportError:
    XGB_AVAILABLE = False

try:
    from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
    SKLEARN_AVAILABLE = True
except ImportError:
    SKLEARN_AVAILABLE = False

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Hyperparameter Tuning")
print("=" * 50)
```

## 1. Chargement des Données

```python
# Chargement des données
features_dir = '../data/features'
symbol = 'BTC-USD'

df = pd.read_csv(os.path.join(features_dir, f'{symbol}_features.csv'), index_col=0, parse_dates=True)
print(f"Données chargées: {len(df)} observations, {len(df.columns)} colonnes")

# Préparation
def prepare_data(df, target_col='close'):
    df = df.copy()
    y = df[target_col].shift(-1)
    X = df.drop(columns=[target_col, 'open', 'high', 'low', 'volume'])
    valid_idx = ~(X.isna().any(axis=1) | y.isna())
    return X[valid_idx], y[valid_idx]

X, y = prepare_data(df)
print(f"Données préparées: {X.shape[0]} échantillons, {X.shape[1]} features")
```

```python
# Split temporel
train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)

X_train = X[:train_size]
y_train = y[:train_size]
X_val = X[train_size:train_size+val_size]
y_val = y[train_size:train_size+val_size]
X_test = X[train_size+val_size:]
y_test = y[train_size+val_size:]

print(f"Train: {len(X_train)}")
print(f"Validation: {len(X_val)}")
print(f"Test: {len(X_test)}")
```

## 2. Grid Search - XGBoost

```python
def grid_search_xgboost(X_train, y_train, X_val, y_val):
    """Grid Search pour XGBoost"""
    if not XGB_AVAILABLE:
        return {}
    
    print("Grid Search pour XGBoost...")
    
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.01, 0.05, 0.1],
        'subsample': [0.7, 0.8, 0.9],
        'colsample_bytree': [0.7, 0.8, 0.9],
    }
    
    model = xgb.XGBRegressor(random_state=42, n_jobs=-1)
    
    tscv = TimeSeriesSplit(n_splits=3)
    
    grid_search = GridSearchCV(
        model,
        param_grid,
        cv=tscv,
        scoring='neg_mean_squared_error',
        n_jobs=-1,
        verbose=1
    )
    
    grid_search.fit(X_train, y_train)
    
    print(f"\nMeilleurs paramètres: {grid_search.best_params_}")
    print(f"Meilleur score: {-grid_search.best_score_:.6f}")
    
    # Évaluation sur validation
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_val)
    
    mse = mean_squared_error(y_val, y_pred)
    mae = mean_absolute_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    
    print(f"\nPerformance sur validation:")
    print(f"  MSE: {mse:.6f}")
    print(f"  MAE: {mae:.6f}")
    print(f"  R2: {r2:.6f}")
    
    return {
        'model': best_model,
        'best_params': grid_search.best_params_,
        'best_score': -grid_search.best_score_,
        'cv_results': grid_search.cv_results_,
    }
```

```python
xgb_results = grid_search_xgboost(X_train, y_train, X_val, y_val)
```

## 3. Random Search - Random Forest

```python
def random_search_random_forest(X_train, y_train, X_val, y_val):
    """Random Search pour Random Forest"""
    if not SKLEARN_AVAILABLE:
        return {}
    
    print("\nRandom Search pour Random Forest...")
    
    param_dist = {
        'n_estimators': [50, 100, 200, 500],
        'max_depth': [5, 10, 15, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['sqrt', 'log2', None],
    }
    
    model = RandomForestRegressor(random_state=42, n_jobs=-1)
    
    tscv = TimeSeriesSplit(n_splits=3)
    
    random_search = RandomizedSearchCV(
        model,
        param_distributions=param_dist,
        n_iter=20,
        cv=tscv,
        scoring='neg_mean_squared_error',
        n_jobs=-1,
        verbose=1,
        random_state=42
    )
    
    random_search.fit(X_train, y_train)
    
    print(f"\nMeilleurs paramètres: {random_search.best_params_}")
    print(f"Meilleur score: {-random_search.best_score_:.6f}")
    
    best_model = random_search.best_estimator_
    y_pred = best_model.predict(X_val)
    
    mse = mean_squared_error(y_val, y_pred)
    mae = mean_absolute_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    
    print(f"\nPerformance sur validation:")
    print(f"  MSE: {mse:.6f}")
    print(f"  MAE: {mae:.6f}")
    print(f"  R2: {r2:.6f}")
    
    return {
        'model': best_model,
        'best_params': random_search.best_params_,
        'best_score': -random_search.best_score_,
    }
```

```python
rf_results = random_search_random_forest(X_train, y_train, X_val, y_val)
```

## 4. Comparaison des Modèles Optimisés

```python
# Test des modèles optimisés
results = {}

if xgb_results:
    y_pred = xgb_results['model'].predict(X_test)
    results['XGBoost'] = {
        'mse': mean_squared_error(y_test, y_pred),
        'mae': mean_absolute_error(y_test, y_pred),
        'r2': r2_score(y_test, y_pred),
        'predictions': y_pred,
    }

if rf_results:
    y_pred = rf_results['model'].predict(X_test)
    results['Random Forest'] = {
        'mse': mean_squared_error(y_test, y_pred),
        'mae': mean_absolute_error(y_test, y_pred),
        'r2': r2_score(y_test, y_pred),
        'predictions': y_pred,
    }

comparison = pd.DataFrame({
    name: {
        'MSE': res['mse'],
        'MAE': res['mae'],
        'R2': res['r2'],
    }
    for name, res in results.items()
}).T

print("Comparaison des modèles optimisés:")
print("=" * 50)
print(comparison.sort_values('R2', ascending=False))
```

## 5. Visualisation des Résultats

```python
# Visualisation des prédictions du meilleur modèle
best_name = comparison['R2'].idxmax()
best_results = results[best_name]

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

ax = axes[0, 0]
ax.plot(y_test.values[-100:], label='Actual', color='blue')
ax.plot(best_results['predictions'][-100:], label='Predicted', color='red', alpha=0.7)
ax.set_title(f'{best_name} - Prédictions (test)')
ax.legend()
ax.grid(True, alpha=0.3)

ax = axes[0, 1]
ax.scatter(y_test.values, best_results['predictions'], alpha=0.5)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Actual vs Predicted')
ax.grid(True, alpha=0.3)

ax = axes[1, 0]
errors = best_results['predictions'] - y_test.values
ax.hist(errors, bins=50, edgecolor='black')
ax.axvline(x=0, color='r', linestyle='--')
ax.set_title('Distribution des erreurs')
ax.set_xlabel('Erreur')
ax.set_ylabel('Fréquence')
ax.grid(True, alpha=0.3)

ax = axes[1, 1]
ax.bar(comparison.index, comparison['R2'])
ax.axhline(y=0, color='black', linestyle='-')
ax.set_title('R2 Score par modèle')
ax.set_xlabel('Modèle')
ax.set_ylabel('R2 Score')
ax.xticks(rotation=45)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Hyperparameter Importance

```python
# Analyse de l'importance des hyperparamètres (XGBoost)
if xgb_results and 'cv_results' in xgb_results:
    cv_results = pd.DataFrame(xgb_results['cv_results'])
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    params = ['param_n_estimators', 'param_max_depth', 'param_learning_rate', 
              'param_subsample', 'param_colsample_bytree']
    
    for i, param in enumerate(params):
        if param in cv_results.columns:
            ax = axes[i // 3, i % 3]
            cv_results.groupby(param)['mean_test_score'].mean().plot(kind='bar', ax=ax)
            ax.set_title(param.replace('param_', ''))
            ax.set_xlabel(param.replace('param_', ''))
            ax.set_ylabel('Mean Test Score')
            ax.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

## 7. Sauvegarde du Meilleur Modèle

```python
# Sélection du meilleur modèle
best_name = comparison['R2'].idxmax()

if best_name == 'XGBoost' and xgb_results:
    best_model = xgb_results['model']
    best_params = xgb_results['best_params']
elif best_name == 'Random Forest' and rf_results:
    best_model = rf_results['model']
    best_params = rf_results['best_params']
else:
    best_model = None
    best_params = {}

print(f"Meilleur modèle: {best_name}")
print(f"Paramètres: {best_params}")
print(f"R2 Score: {comparison.loc[best_name, 'R2']:.6f}")

# Sauvegarde
models_dir = '../models'
os.makedirs(models_dir, exist_ok=True)

if best_model:
    model_path = os.path.join(models_dir, f'{symbol}_{best_name.lower()}_tuned.pkl')
    with open(model_path, 'wb') as f:
        pickle.dump(best_model, f)
    print(f"Modèle sauvegardé: {model_path}")

# Sauvegarde des paramètres
params_path = os.path.join(models_dir, f'{symbol}_best_params.pkl')
with open(params_path, 'wb') as f:
    pickle.dump({
        'model': best_name,
        'params': best_params,
        'score': comparison.loc[best_name, 'R2'],
        'timestamp': datetime.now().isoformat(),
    }, f)
print(f"Paramètres sauvegardés: {params_path}")

print("\n" + "=" * 50)
print("Hyperparameter tuning terminé!")
print("=" * 50)
```
