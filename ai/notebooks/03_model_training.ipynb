```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pickle
import os
import warnings
warnings.filterwarnings('ignore')

# Scikit-learn
from sklearn.model_selection import train_test_split, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Modèles
try:
    import xgboost as xgb
    XGB_AVAILABLE = True
except ImportError:
    XGB_AVAILABLE = False

try:
    from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
    from sklearn.linear_model import LinearRegression, Ridge, Lasso
    SKLEARN_AVAILABLE = True
except ImportError:
    SKLEARN_AVAILABLE = False

try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from torch.utils.data import DataLoader, TensorDataset
    TORCH_AVAILABLE = True
except ImportError:
    TORCH_AVAILABLE = False

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Model Training")
print("=" * 50)
print(f"XGBoost disponible: {XGB_AVAILABLE}")
print(f"Scikit-learn disponible: {SKLEARN_AVAILABLE}")
print(f"PyTorch disponible: {TORCH_AVAILABLE}")
```

## 1. Chargement des Features

```python
# Chargement des données
features_dir = '../data/features'

data = {}
for file in os.listdir(features_dir):
    if file.endswith('_features.csv'):
        symbol = file.replace('_features.csv', '')
        df = pd.read_csv(os.path.join(features_dir, file), index_col=0, parse_dates=True)
        data[symbol] = df
        print(f"Chargé: {symbol} ({len(df)} observations, {len(df.columns)} features)")

# Chargement des scalers
scaler_path = os.path.join(features_dir, 'scalers.pkl')
with open(scaler_path, 'rb') as f:
    scalers = pickle.load(f)
```

## 2. Préparation des Données

```python
def prepare_data(df, target_col='close', lookback=0):
    """Prépare les données pour l'entraînement"""
    df = df.copy()
    
    # Target
    y = df[target_col].shift(-1)
    X = df.drop(columns=[target_col, 'open', 'high', 'low', 'volume'])
    
    # Drop NaN
    valid_idx = ~(X.isna().any(axis=1) | y.isna())
    X = X[valid_idx]
    y = y[valid_idx]
    
    return X, y

# Préparation
symbol = 'BTC-USD'
df = data[symbol]
X, y = prepare_data(df)

print(f"Données préparées: {X.shape[0]} échantillons, {X.shape[1]} features")
print(f"Target: {y.shape[0]} valeurs")
```

```python
# Split temporel
split_idx = int(len(X) * 0.8)

X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"Train: {X_train.shape[0]} samples")
print(f"Test: {X_test.shape[0]} samples")
print(f"Train dates: {X_train.index[0]} -> {X_train.index[-1]}")
print(f"Test dates: {X_test.index[0]} -> {X_test.index[-1]}")
```

## 3. Modèles ML (Scikit-learn)

```python
def train_sklearn_models(X_train, y_train, X_test, y_test):
    """Entraîne les modèles scikit-learn"""
    if not SKLEARN_AVAILABLE:
        return {}
    
    models = {
        'Linear Regression': LinearRegression(),
        'Ridge': Ridge(alpha=1.0),
        'Lasso': Lasso(alpha=0.01),
        'Random Forest': RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42),
        'Gradient Boosting': GradientBoostingRegressor(n_estimators=100, max_depth=5, random_state=42),
    }
    
    results = {}
    
    for name, model in models.items():
        print(f"Entraînement de {name}...")
        model.fit(X_train, y_train)
        
        y_pred = model.predict(X_test)
        
        mse = mean_squared_error(y_test, y_pred)
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        
        results[name] = {
            'model': model,
            'predictions': y_pred,
            'mse': mse,
            'mae': mae,
            'r2': r2,
        }
        
        print(f"  MSE: {mse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}")
    
    return results

ml_results = train_sklearn_models(X_train, y_train, X_test, y_test)
```

## 4. Modèle XGBoost

```python
def train_xgboost_model(X_train, y_train, X_test, y_test):
    """Entraîne le modèle XGBoost"""
    if not XGB_AVAILABLE:
        return {}
    
    print("Entraînement de XGBoost...")
    
    model = xgb.XGBRegressor(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42
    )
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"  MSE: {mse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}")
    
    return {
        'model': model,
        'predictions': y_pred,
        'mse': mse,
        'mae': mae,
        'r2': r2,
        'feature_importance': model.feature_importances_,
    }
```

```python
xgb_result = train_xgboost_model(X_train, y_train, X_test, y_test)
```

## 5. Modèle LSTM (PyTorch)

```python
def train_lstm_model(X_train, y_train, X_test, y_test, sequence_length=24):
    """Entraîne le modèle LSTM"""
    if not TORCH_AVAILABLE:
        return {}
    
    print("Entraînement de LSTM...")
    
    class LSTMModel(nn.Module):
        def __init__(self, input_size, hidden_size=64, num_layers=2):
            super().__init__()
            self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
            self.fc = nn.Linear(hidden_size, 1)
        
        def forward(self, x):
            lstm_out, _ = self.lstm(x)
            return self.fc(lstm_out[:, -1, :])
    
    # Création des séquences
    def create_sequences(X, y, seq_len):
        X_seq, y_seq = [], []
        for i in range(len(X) - seq_len):
            X_seq.append(X[i:i+seq_len])
            y_seq.append(y[i+seq_len])
        return np.array(X_seq), np.array(y_seq)
    
    X_train_seq, y_train_seq = create_sequences(X_train.values, y_train.values, sequence_length)
    X_test_seq, y_test_seq = create_sequences(X_test.values, y_test.values, sequence_length)
    
    # Convertir en tenseurs
    X_train_t = torch.FloatTensor(X_train_seq)
    y_train_t = torch.FloatTensor(y_train_seq).reshape(-1, 1)
    X_test_t = torch.FloatTensor(X_test_seq)
    y_test_t = torch.FloatTensor(y_test_seq).reshape(-1, 1)
    
    # DataLoader
    train_dataset = TensorDataset(X_train_t, y_train_t)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    
    # Modèle
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = LSTMModel(X_train_seq.shape[2]).to(device)
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    
    # Entraînement
    for epoch in range(50):
        model.train()
        epoch_loss = 0
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            output = model(batch_X)
            loss = criterion(output, batch_y)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
        
        if epoch % 10 == 0:
            print(f"  Epoch {epoch}: Loss = {epoch_loss/len(train_loader):.6f}")
    
    # Prédiction
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test_t.to(device)).cpu().numpy()
    
    mse = mean_squared_error(y_test_seq, y_pred)
    mae = mean_absolute_error(y_test_seq, y_pred)
    r2 = r2_score(y_test_seq, y_pred)
    
    print(f"  MSE: {mse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}")
    
    return {
        'model': model,
        'predictions': y_pred.flatten(),
        'mse': mse,
        'mae': mae,
        'r2': r2,
    }
```

```python
lstm_result = train_lstm_model(X_train, y_train, X_test, y_test)
```

## 6. Comparaison des Modèles

```python
# Comparaison des résultats
all_results = {}
if ml_results:
    all_results.update(ml_results)
if xgb_result:
    all_results['XGBoost'] = xgb_result
if lstm_result:
    all_results['LSTM'] = lstm_result

comparison = pd.DataFrame({
    name: {
        'MSE': res['mse'],
        'MAE': res['mae'],
        'R2': res['r2'],
    }
    for name, res in all_results.items()
}).T

print("Comparaison des modèles:")
print("=" * 50)
print(comparison.sort_values('R2', ascending=False))
```

```python
# Visualisation des prédictions
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Meilleur modèle
best_name = comparison['R2'].idxmax()
best_result = all_results[best_name]

ax = axes[0, 0]
ax.plot(y_test.values[-100:], label='Actual', color='blue')
ax.plot(best_result['predictions'][-100:], label='Predicted', color='red', alpha=0.7)
ax.set_title(f'{best_name} - Prédictions')
ax.legend()
ax.grid(True, alpha=0.3)

# Scatter plot
ax = axes[0, 1]
ax.scatter(y_test.values, best_result['predictions'], alpha=0.5)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Actual vs Predicted')
ax.grid(True, alpha=0.3)

# Erreurs
ax = axes[1, 0]
errors = best_result['predictions'] - y_test.values
ax.hist(errors, bins=50, edgecolor='black')
ax.axvline(x=0, color='r', linestyle='--')
ax.set_title('Distribution des erreurs')
ax.set_xlabel('Erreur')
ax.set_ylabel('Fréquence')
ax.grid(True, alpha=0.3)

# Résumé des performances
ax = axes[1, 1]
ax.bar(comparison.index, comparison['R2'])
ax.axhline(y=0, color='black', linestyle='-')
ax.set_title('R2 Score par modèle')
ax.set_xlabel('Modèle')
ax.set_ylabel('R2 Score')
ax.xticks(rotation=45)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 7. Feature Importance

```python
# Feature importance du meilleur modèle
if best_name == 'XGBoost' and xgb_result:
    importance = xgb_result['feature_importance']
elif 'Random Forest' in all_results and best_name == 'Random Forest':
    importance = all_results['Random Forest']['model'].feature_importances_
else:
    importance = None

if importance is not None:
    feature_names = X.columns
    importance_df = pd.DataFrame({
        'feature': feature_names,
        'importance': importance
    }).sort_values('importance', ascending=False)

    plt.figure(figsize=(12, 8))
    importance_df.head(20).plot(kind='barh', x='feature', y='importance')
    plt.title(f'Top 20 Features - {best_name}')
    plt.xlabel('Importance')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

    print("Top 10 features:")
    print(importance_df.head(10))
```

## 8. Sauvegarde du Modèle

```python
# Sauvegarde du meilleur modèle
models_dir = '../models'
os.makedirs(models_dir, exist_ok=True)

best_model = all_results[best_name]['model']
model_path = os.path.join(models_dir, f'{symbol}_{best_name.lower()}.pkl')

with open(model_path, 'wb') as f:
    pickle.dump(best_model, f)

print(f"Modèle sauvegardé: {model_path}")

# Sauvegarde des métriques
metrics = {
    'model': best_name,
    'mse': all_results[best_name]['mse'],
    'mae': all_results[best_name]['mae'],
    'r2': all_results[best_name]['r2'],
    'train_samples': len(X_train),
    'test_samples': len(X_test),
    'features': list(X.columns),
}

metrics_path = os.path.join(models_dir, f'{symbol}_metrics.pkl')
with open(metrics_path, 'wb') as f:
    pickle.dump(metrics, f)

print(f"Métriques sauvegardées: {metrics_path}")

# Sauvegarde des scalers
scaler_path = os.path.join(models_dir, f'{symbol}_scaler.pkl')
with open(scaler_path, 'wb') as f:
    pickle.dump(scalers[symbol], f)

print(f"Scaler sauvegardé: {scaler_path}")

print("\n" + "=" * 50)
print(f"Meilleur modèle: {best_name}")
print(f"R2 Score: {all_results[best_name]['r2']:.4f}")
print("Model training terminé!")
print("=" * 50)
```
