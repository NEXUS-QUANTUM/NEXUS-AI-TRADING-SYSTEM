```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Data Exploration")
print("=" * 50)
```

## 1. Chargement des Données

```python
# Simulation de données de marché
def generate_market_data(days=365, symbols=['BTC-USD', 'ETH-USD', 'SOL-USD']):
    """Génère des données de marché synthétiques"""
    dates = pd.date_range(
        start=datetime.now() - timedelta(days=days),
        end=datetime.now(),
        freq='D'
    )
    
    data = {}
    base_prices = {'BTC-USD': 50000, 'ETH-USD': 3000, 'SOL-USD': 100}
    
    for symbol in symbols:
        n = len(dates)
        base_price = base_prices.get(symbol, 1000)
        drift = np.random.normal(0.0001, 0.02, n)
        price = base_price * np.cumprod(1 + drift)
        
        df = pd.DataFrame({
            'date': dates,
            'open': price * (1 + np.random.uniform(-0.01, 0.01, n)),
            'high': price * (1 + np.random.uniform(0, 0.02, n)),
            'low': price * (1 + np.random.uniform(-0.02, 0, n)),
            'close': price,
            'volume': np.random.uniform(1000, 10000, n),
        })
        df.set_index('date', inplace=True)
        data[symbol] = df
    
    return data

# Génération des données
data = generate_market_data(days=365)
print(f"Données générées pour {len(data)} symboles")
for symbol, df in data.items():
    print(f"  {symbol}: {len(df)} observations")
```

## 2. Analyse Exploratoire

```python
# Aperçu des données
symbol = 'BTC-USD'
df = data[symbol]
print(f"Aperçu des données pour {symbol}")
print("=" * 50)
print(df.head())
print("\nStatistiques descriptives:")
print(df.describe())
```

```python
# Visualisation des prix
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Prix
ax = axes[0, 0]
for symbol, df in data.items():
    ax.plot(df.index, df['close'], label=symbol, alpha=0.7)
ax.set_title('Évolution des prix')
ax.set_xlabel('Date')
ax.set_ylabel('Prix')
ax.legend()
ax.grid(True, alpha=0.3)

# Rendements
ax = axes[0, 1]
for symbol, df in data.items():
    returns = df['close'].pct_change().dropna()
    ax.hist(returns, bins=50, alpha=0.5, label=symbol)
ax.set_title('Distribution des rendements')
ax.set_xlabel('Rendement')
ax.set_ylabel('Fréquence')
ax.legend()
ax.grid(True, alpha=0.3)

# Boxplot
ax = axes[1, 0]
returns_data = []
labels = []
for symbol, df in data.items():
    returns = df['close'].pct_change().dropna()
    returns_data.append(returns)
    labels.append(symbol)
ax.boxplot(returns_data, labels=labels)
ax.set_title('Boxplot des rendements')
ax.set_ylabel('Rendement')
ax.grid(True, alpha=0.3)

# Volume
ax = axes[1, 1]
for symbol, df in data.items():
    ax.plot(df.index, df['volume'], label=symbol, alpha=0.7)
ax.set_title('Volume des transactions')
ax.set_xlabel('Date')
ax.set_ylabel('Volume')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Indicateurs Techniques

```python
# Calcul des indicateurs techniques
def add_technical_indicators(df):
    """Ajoute des indicateurs techniques"""
    df = df.copy()
    
    # Moyennes mobiles
    df['ma_7'] = df['close'].rolling(7).mean()
    df['ma_20'] = df['close'].rolling(20).mean()
    df['ma_50'] = df['close'].rolling(50).mean()
    
    # RSI
    delta = df['close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    df['rsi'] = 100 - (100 / (1 + rs))
    
    # MACD
    exp1 = df['close'].ewm(span=12, adjust=False).mean()
    exp2 = df['close'].ewm(span=26, adjust=False).mean()
    df['macd'] = exp1 - exp2
    df['signal'] = df['macd'].ewm(span=9, adjust=False).mean()
    
    # Bollinger Bands
    df['bb_middle'] = df['close'].rolling(20).mean()
    bb_std = df['close'].rolling(20).std()
    df['bb_upper'] = df['bb_middle'] + 2 * bb_std
    df['bb_lower'] = df['bb_middle'] - 2 * bb_std
    
    # ATR
    high_low = df['high'] - df['low']
    high_close = np.abs(df['high'] - df['close'].shift())
    low_close = np.abs(df['low'] - df['close'].shift())
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    df['atr'] = tr.rolling(14).mean()
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_technical_indicators(df)

print("Indicateurs techniques ajoutés")
print(data['BTC-USD'].head())
```

```python
# Visualisation des indicateurs
df = data['BTC-USD']

fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# Prix et moyennes mobiles
ax = axes[0]
ax.plot(df.index, df['close'], label='Close', color='blue', linewidth=1)
ax.plot(df.index, df['ma_20'], label='MA 20', color='orange', linestyle='--')
ax.plot(df.index, df['ma_50'], label='MA 50', color='green', linestyle='--')
ax.fill_between(df.index, df['bb_lower'], df['bb_upper'], alpha=0.2, color='gray')
ax.set_title('Prix et indicateurs - BTC-USD')
ax.legend()
ax.grid(True, alpha=0.3)

# RSI
ax = axes[1]
ax.plot(df.index, df['rsi'], label='RSI', color='purple')
ax.axhline(y=70, color='r', linestyle='--', alpha=0.5)
ax.axhline(y=30, color='g', linestyle='--', alpha=0.5)
ax.fill_between(df.index, 30, 70, alpha=0.1, color='gray')
ax.set_ylim(0, 100)
ax.set_title('RSI')
ax.legend()
ax.grid(True, alpha=0.3)

# MACD
ax = axes[2]
ax.plot(df.index, df['macd'], label='MACD', color='blue')
ax.plot(df.index, df['signal'], label='Signal', color='red')
ax.bar(df.index, df['macd'] - df['signal'], label='Histogram', alpha=0.3)
ax.set_title('MACD')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Statistiques Avancées

```python
# Analyse des rendements
for symbol, df in data.items():
    returns = df['close'].pct_change().dropna()
    print(f"\n{symbol}")
    print("-" * 30)
    print(f"  Moyenne: {returns.mean():.6f}")
    print(f"  Écart-type: {returns.std():.6f}")
    print(f"  Skewness: {returns.skew():.4f}")
    print(f"  Kurtosis: {returns.kurtosis():.4f}")
    print(f"  Sharpe ratio: {returns.mean() / returns.std() * np.sqrt(252):.4f}")
```

```python
# Drawdown
def calculate_drawdown(prices):
    """Calcule le drawdown"""
    peak = prices.expanding().max()
    drawdown = (peak - prices) / peak
    return drawdown

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Drawdown
ax = axes[0]
for symbol, df in data.items():
    dd = calculate_drawdown(df['close'])
    ax.plot(df.index, dd * 100, label=symbol, alpha=0.7)
ax.set_title('Drawdown')
ax.set_xlabel('Date')
ax.set_ylabel('Drawdown (%)')
ax.legend()
ax.grid(True, alpha=0.3)

# Volatilité glissante
ax = axes[1]
for symbol, df in data.items():
    vol = df['close'].pct_change().rolling(20).std() * np.sqrt(252)
    ax.plot(df.index, vol * 100, label=symbol, alpha=0.7)
ax.set_title('Volatilité glissante (20 jours)')
ax.set_xlabel('Date')
ax.set_ylabel('Volatilité annualisée (%)')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Détection de Points de Retournement

```python
# Détection des sommets et creux
def find_peaks_troughs(prices, lookback=5):
    """Trouve les sommets et creux locaux"""
    peaks = []
    troughs = []
    
    for i in range(lookback, len(prices) - lookback):
        if prices[i] == max(prices[i-lookback:i+lookback+1]):
            peaks.append((i, prices[i]))
        if prices[i] == min(prices[i-lookback:i+lookback+1]):
            troughs.append((i, prices[i]))
    
    return peaks, troughs

df = data['BTC-USD']
peaks, troughs = find_peaks_troughs(df['close'].values, lookback=10)

plt.figure(figsize=(15, 6))
plt.plot(df.index, df['close'], label='Price', color='blue')

if peaks:
    plt.scatter([df.index[i] for i, _ in peaks], [p for _, p in peaks], 
                color='red', s=50, label='Peaks', zorder=5)
if troughs:
    plt.scatter([df.index[i] for i, _ in troughs], [p for _, p in troughs], 
                color='green', s=50, label='Troughs', zorder=5)

plt.title('Détection des sommets et creux - BTC-USD')
plt.xlabel('Date')
plt.ylabel('Prix')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
