```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pickle
import os
import warnings
warnings.filterwarnings('ignore')

# Scikit-learn
from sklearn.metrics import (
    mean_squared_error, mean_absolute_error, r2_score,
    mean_absolute_percentage_error, explained_variance_score
)

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Model Evaluation")
print("=" * 50)
```

## 1. Chargement des Données et Modèles

```python
# Chargement des données
features_dir = '../data/features'
symbol = 'BTC-USD'

df = pd.read_csv(os.path.join(features_dir, f'{symbol}_features.csv'), index_col=0, parse_dates=True)
print(f"Données chargées: {len(df)} observations")

# Préparation
def prepare_data(df, target_col='close'):
    df = df.copy()
    y = df[target_col].shift(-1)
    X = df.drop(columns=[target_col, 'open', 'high', 'low', 'volume'])
    valid_idx = ~(X.isna().any(axis=1) | y.isna())
    return X[valid_idx], y[valid_idx]

X, y = prepare_data(df)

# Split
split_idx = int(len(X) * 0.8)
X_train, X_test = X[:split_idx], X[split_idx:]
y_train, y_test = y[:split_idx], y[split_idx:]

print(f"Train: {len(X_train)}, Test: {len(X_test)}")
```

```python
# Chargement des modèles
models_dir = '../models'
models = {}

for file in os.listdir(models_dir):
    if file.startswith(symbol) and file.endswith('.pkl') and 'scaler' not in file and 'metrics' not in file and 'params' not in file:
        name = file.replace(f'{symbol}_', '').replace('.pkl', '')
        with open(os.path.join(models_dir, file), 'rb') as f:
            models[name] = pickle.load(f)
        print(f"Modèle chargé: {name}")

if not models:
    print("Aucun modèle trouvé")
```

## 2. Métriques de Performance

```python
def evaluate_model(model, X_test, y_test, name):
    """Évalue un modèle avec plusieurs métriques"""
    y_pred = model.predict(X_test)
    
    metrics = {
        'MSE': mean_squared_error(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred),
        'MAPE': mean_absolute_percentage_error(y_test, y_pred) * 100,
        'R2': r2_score(y_test, y_pred),
        'Explained Variance': explained_variance_score(y_test, y_pred),
    }
    
    return metrics, y_pred

results = {}
predictions = {}

for name, model in models.items():
    if hasattr(model, 'predict'):
        metrics, y_pred = evaluate_model(model, X_test, y_test, name)
        results[name] = metrics
        predictions[name] = y_pred
        print(f"\n{name}:")
        for key, value in metrics.items():
            print(f"  {key}: {value:.6f}")
```

```python
# Comparaison des modèles
comparison_df = pd.DataFrame(results).T
comparison_df = comparison_df.sort_values('R2', ascending=False)

print("\n" + "=" * 50)
print("Comparaison des modèles:")
print("=" * 50)
print(comparison_df)
```

## 3. Analyse des Résidus

```python
# Analyse des résidus pour le meilleur modèle
best_name = comparison_df.index[0]
best_pred = predictions[best_name]
residuals = y_test.values - best_pred

fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Distribution des résidus
ax = axes[0, 0]
ax.hist(residuals, bins=50, edgecolor='black', alpha=0.7)
ax.axvline(x=0, color='r', linestyle='--')
ax.set_title(f'{best_name} - Distribution des résidus')
ax.set_xlabel('Résidu')
ax.set_ylabel('Fréquence')
ax.grid(True, alpha=0.3)

# Résidus vs Prédictions
ax = axes[0, 1]
ax.scatter(best_pred, residuals, alpha=0.5)
ax.axhline(y=0, color='r', linestyle='--')
ax.set_title('Résidus vs Prédictions')
ax.set_xlabel('Prédictions')
ax.set_ylabel('Résidus')
ax.grid(True, alpha=0.3)

# Q-Q Plot
ax = axes[1, 0]
from scipy import stats
stats.probplot(residuals, dist="norm", plot=ax)
ax.set_title('Q-Q Plot des résidus')
ax.grid(True, alpha=0.3)

# AC et PAC
ax = axes[1, 1]
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(residuals, lags=20, ax=ax)
ax.set_title('ACF des résidus')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Erreurs de Prédiction

```python
# Analyse des erreurs
errors = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': best_pred,
    'Error': residuals,
    'Absolute Error': np.abs(residuals),
    'Percentage Error': np.abs(residuals / y_test.values) * 100,
}, index=X_test.index)

print("Statistiques des erreurs:")
print("=" * 50)
print(errors.describe())
```

```python
# Visualisation des erreurs
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

ax = axes[0, 0]
ax.plot(errors.index, errors['Error'], alpha=0.7)
ax.axhline(y=0, color='r', linestyle='--')
ax.set_title('Évolution des erreurs')
ax.set_xlabel('Temps')
ax.set_ylabel('Erreur')
ax.grid(True, alpha=0.3)

ax = axes[0, 1]
ax.scatter(errors['Actual'], errors['Absolute Error'], alpha=0.5)
ax.set_title('Erreur absolue vs Valeur réelle')
ax.set_xlabel('Valeur réelle')
ax.set_ylabel('Erreur absolue')
ax.grid(True, alpha=0.3)

ax = axes[1, 0]
ax.hist(errors['Percentage Error'], bins=50, edgecolor='black', alpha=0.7)
ax.set_title('Distribution des erreurs en pourcentage')
ax.set_xlabel('Erreur (%)')
ax.set_ylabel('Fréquence')
ax.grid(True, alpha=0.3)

ax = axes[1, 1]
ax.boxplot([errors['Error']])
ax.set_title('Boxplot des erreurs')
ax.set_ylabel('Erreur')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 5. Erreurs par Période

```python
# Analyse des erreurs par période
errors['Year'] = errors.index.year
errors['Month'] = errors.index.month
errors['Quarter'] = errors.index.quarter

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Par année
ax = axes[0]
yearly_mae = errors.groupby('Year')['Absolute Error'].mean()
yearly_mae.plot(kind='bar', ax=ax)
ax.set_title('MAE par année')
ax.set_xlabel('Année')
ax.set_ylabel('MAE')
ax.grid(True, alpha=0.3)

# Par mois
ax = axes[1]
monthly_mae = errors.groupby('Month')['Absolute Error'].mean()
monthly_mae.plot(kind='bar', ax=ax)
ax.set_title('MAE par mois')
ax.set_xlabel('Mois')
ax.set_ylabel('MAE')
ax.grid(True, alpha=0.3)

# Par trimestre
ax = axes[2]
quarterly_mae = errors.groupby('Quarter')['Absolute Error'].mean()
quarterly_mae.plot(kind='bar', ax=ax)
ax.set_title('MAE par trimestre')
ax.set_xlabel('Trimestre')
ax.set_ylabel('MAE')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Robustesse du Modèle

```python
# Test de robustesse
from sklearn.model_selection import TimeSeriesSplit

def cross_validate_model(model, X, y, n_splits=5):
    """Validation croisée temporelle"""
    tscv = TimeSeriesSplit(n_splits=n_splits)
    scores = []
    
    for train_idx, val_idx in tscv.split(X):
        X_train_cv, X_val_cv = X.iloc[train_idx], X.iloc[val_idx]
        y_train_cv, y_val_cv = y.iloc[train_idx], y.iloc[val_idx]
        
        model.fit(X_train_cv, y_train_cv)
        y_pred = model.predict(X_val_cv)
        scores.append(r2_score(y_val_cv, y_pred))
    
    return np.mean(scores), np.std(scores)

best_model = models[best_name]
mean_score, std_score = cross_validate_model(best_model, X, y)

print(f"Validation croisée pour {best_name}:")
print("=" * 50)
print(f"R2 moyen: {mean_score:.6f}")
print(f"Écart-type: {std_score:.6f}")
print(f"Stabilité: {(1 - std_score/mean_score)*100:.2f}%")
```

## 7. Sauvegarde des Résultats

```python
# Sauvegarde des résultats
results_dir = '../results'
os.makedirs(results_dir, exist_ok=True)

# Métriques
metrics_path = os.path.join(results_dir, f'{symbol}_evaluation_metrics.csv')
comparison_df.to_csv(metrics_path)
print(f"Métriques sauvegardées: {metrics_path}")

# Erreurs
errors_path = os.path.join(results_dir, f'{symbol}_errors.csv')
errors.to_csv(errors_path)
print(f"Erreurs sauvegardées: {errors_path}")

# Résumé
summary = {
    'best_model': best_name,
    'r2_score': comparison_df.loc[best_name, 'R2'],
    'rmse': comparison_df.loc[best_name, 'RMSE'],
    'mae': comparison_df.loc[best_name, 'MAE'],
    'cv_mean': mean_score,
    'cv_std': std_score,
    'timestamp': datetime.now().isoformat(),
}

summary_path = os.path.join(results_dir, f'{symbol}_evaluation_summary.pkl')
with open(summary_path, 'wb') as f:
    pickle.dump(summary, f)
print(f"Résumé sauvegardé: {summary_path}")

print("\n" + "=" * 50)
print("Évaluation du modèle terminée!")
print("=" * 50)
```
