Ce notebook effectue le backtest des stratégies de trading.

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import pickle
import os
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Backtesting")
print("=" * 50)
```

## 1. Chargement des Données et Modèles

```python
# Chargement des données
data_dir = '../data/features'

data = {}
for file in os.listdir(data_dir):
    if file.endswith('_features.csv'):
        symbol = file.replace('_features.csv', '')
        df = pd.read_csv(os.path.join(data_dir, file), index_col=0, parse_dates=True)
        data[symbol] = df
        print(f"Chargé: {symbol} ({len(df)} observations)")

# Chargement du modèle
models_dir = '../models'
symbol = 'BTC-USD'

model_files = [f for f in os.listdir(models_dir) if f.startswith(symbol) and f.endswith('.pkl') and 'scaler' not in f and 'metrics' not in f]
if model_files:
    model_path = os.path.join(models_dir, model_files[0])
    with open(model_path, 'rb') as f:
        model = pickle.load(f)
    print(f"Modèle chargé: {model_path}")
else:
    print("Aucun modèle trouvé")
    model = None
```

## 2. Stratégie de Trading

```python
class TradingStrategy:
    """Stratégie de trading basée sur les prédictions du modèle"""
    
    def __init__(self, model, scaler, threshold=0.005):
        self.model = model
        self.scaler = scaler
        self.threshold = threshold
        self.position = 0
        self.trades = []
        self.equity = [10000.0]
        
    def get_features(self, df):
        """Extrait les features"""
        exclude = ['open', 'high', 'low', 'close', 'volume']
        features = df.drop(columns=[c for c in exclude if c in df.columns], errors='ignore')
        return features
    
    def backtest(self, df):
        """Effectue le backtest"""
        df = df.copy()
        
        for i in range(len(df)):
            current_data = df.iloc[:i+1]
            if len(current_data) < 50:
                continue
            
            try:
                features = self.get_features(current_data)
                features_scaled = self.scaler.transform(features.iloc[-1:].values)
                
                if self.model is not None:
                    pred = self.model.predict(features_scaled)[0]
                    signal = pred - current_data['close'].iloc[-1]
                    
                    # Signal de trading
                    if signal > self.threshold and self.position == 0:
                        self.position = 1
                        self.trades.append({
                            'entry_time': current_data.index[-1],
                            'entry_price': current_data['close'].iloc[-1],
                            'position': 1
                        })
                    elif signal < -self.threshold and self.position == 1:
                        self.position = 0
                        if self.trades:
                            trade = self.trades[-1]
                            trade['exit_time'] = current_data.index[-1]
                            trade['exit_price'] = current_data['close'].iloc[-1]
                            trade['pnl'] = trade['exit_price'] - trade['entry_price']
                            self.equity.append(self.equity[-1] + trade['pnl'])
                else:
                    # Stratégie buy and hold
                    self.equity.append(self.equity[-1] * (1 + current_data['close'].pct_change().iloc[-1]))
                    
            except Exception as e:
                continue
        
        return self
    
    def get_performance(self):
        """Retourne les performances"""
        returns = np.diff(self.equity) / self.equity[:-1]
        
        total_return = (self.equity[-1] / self.equity[0]) - 1
        sharpe = np.mean(returns) / np.std(returns) * np.sqrt(252) if len(returns) > 1 else 0
        max_drawdown = self._calculate_max_drawdown()
        
        return {
            'total_return': total_return * 100,
            'sharpe_ratio': sharpe,
            'max_drawdown': max_drawdown * 100,
            'trades': len(self.trades),
            'equity': self.equity,
        }
    
    def _calculate_max_drawdown(self):
        """Calcule le drawdown maximum"""
        peak = np.maximum.accumulate(self.equity)
        drawdown = (peak - self.equity) / peak
        return np.max(drawdown) if len(drawdown) > 0 else 0
```

```python
# Chargement du scaler
scaler_path = os.path.join(models_dir, f'{symbol}_scaler.pkl')
with open(scaler_path, 'rb') as f:
    scaler = pickle.load(f)

# Création de la stratégie
strategy = TradingStrategy(model, scaler, threshold=0.005)
strategy.backtest(data[symbol])

performance = strategy.get_performance()

print("Performances du backtest:")
print("=" * 50)
print(f"Rendement total: {performance['total_return']:.2f}%")
print(f"Ratio de Sharpe: {performance['sharpe_ratio']:.4f}")
print(f"Drawdown maximum: {performance['max_drawdown']:.2f}%")
print(f"Nombre de trades: {performance['trades']}")
```

## 3. Visualisation des Résultats

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Courbe d'équité
ax = axes[0, 0]
ax.plot(strategy.equity, label='Equity Curve', color='blue')
ax.set_title('Courbe d\'équité')
ax.set_xlabel('Temps')
ax.set_ylabel('Équité ($)')
ax.axhline(y=10000, color='r', linestyle='--', alpha=0.5)
ax.grid(True, alpha=0.3)

# Drawdown
ax = axes[0, 1]
peak = np.maximum.accumulate(strategy.equity)
drawdown = (peak - strategy.equity) / peak * 100
ax.fill_between(range(len(drawdown)), 0, drawdown, color='red', alpha=0.3)
ax.set_title('Drawdown')
ax.set_xlabel('Temps')
ax.set_ylabel('Drawdown (%)')
ax.grid(True, alpha=0.3)

# Returns
ax = axes[1, 0]
returns = np.diff(strategy.equity) / strategy.equity[:-1] * 100
ax.bar(range(len(returns)), returns, alpha=0.5)
ax.axhline(y=0, color='black', linestyle='-')
ax.set_title('Rendements journaliers')
ax.set_xlabel('Temps')
ax.set_ylabel('Rendement (%)')
ax.grid(True, alpha=0.3)

# Distribution des returns
ax = axes[1, 1]
ax.hist(returns, bins=30, edgecolor='black', alpha=0.7)
ax.axvline(x=np.mean(returns), color='r', linestyle='--', label=f'Moyenne: {np.mean(returns):.2f}%')
ax.set_title('Distribution des rendements')
ax.set_xlabel('Rendement (%)')
ax.set_ylabel('Fréquence')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 4. Analyse des Trades

```python
# Analyse des trades
if strategy.trades:
    trades_df = pd.DataFrame(strategy.trades)
    
    print("Analyse des trades:")
    print("=" * 50)
    print(f"Nombre de trades: {len(trades_df)}")
    
    if 'pnl' in trades_df.columns:
        winning_trades = trades_df[trades_df['pnl'] > 0]
        losing_trades = trades_df[trades_df['pnl'] < 0]
        
        print(f"Trades gagnants: {len(winning_trades)} ({len(winning_trades)/len(trades_df)*100:.1f}%)")
        print(f"Trades perdants: {len(losing_trades)} ({len(losing_trades)/len(trades_df)*100:.1f}%)")
        
        if len(winning_trades) > 0:
            print(f"Gain moyen: {winning_trades['pnl'].mean():.2f}$")
        if len(losing_trades) > 0:
            print(f"Perte moyenne: {losing_trades['pnl'].mean():.2f}$")
        
        profit_factor = abs(winning_trades['pnl'].sum() / losing_trades['pnl'].sum()) if len(losing_trades) > 0 else float('inf')
        print(f"Profit factor: {profit_factor:.2f}")
        
        # Visualisation
        fig, axes = plt.subplots(1, 2, figsize=(12, 5))
        
        ax = axes[0]
        trades_df['pnl'].hist(bins=20, edgecolor='black', alpha=0.7)
        ax.axvline(x=0, color='r', linestyle='--')
        ax.set_title('Distribution des P&L')
        ax.set_xlabel('P&L ($)')
        ax.set_ylabel('Fréquence')
        ax.grid(True, alpha=0.3)
        
        ax = axes[1]
        trades_df['pnl'].cumsum().plot()
        ax.set_title('P&L cumulé')
        ax.set_xlabel('Trade')
        ax.set_ylabel('P&L cumulé ($)')
        ax.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()
```

## 5. Comparaison des Stratégies

```python
# Comparaison avec buy and hold
buy_hold_strategy = TradingStrategy(None, scaler, 0)
buy_hold_strategy.backtest(data[symbol])
buy_hold_performance = buy_hold_strategy.get_performance()

# Comparaison
comparison = pd.DataFrame({
    'Métrique': ['Rendement total (%)', 'Ratio de Sharpe', 'Drawdown max (%)', 'Trades'],
    'Stratégie ML': [
        performance['total_return'],
        performance['sharpe_ratio'],
        performance['max_drawdown'],
        performance['trades']
    ],
    'Buy & Hold': [
        buy_hold_performance['total_return'],
        buy_hold_performance['sharpe_ratio'],
        buy_hold_performance['max_drawdown'],
        buy_hold_performance['trades']
    ]
})

print("Comparaison des stratégies:")
print("=" * 50)
print(comparison.to_string(index=False))
```

```python
# Visualisation comparative
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

ax = axes[0]
ax.plot(strategy.equity, label='Stratégie ML', color='blue')
ax.plot(buy_hold_strategy.equity, label='Buy & Hold', color='green')
ax.set_title('Comparaison des courbes d\'équité')
ax.set_xlabel('Temps')
ax.set_ylabel('Équité ($)')
ax.legend()
ax.grid(True, alpha=0.3)

ax = axes[1]
ax.bar(comparison['Métrique'], comparison['Stratégie ML'], label='Stratégie ML', alpha=0.7)
ax.bar(comparison['Métrique'], comparison['Buy & Hold'], label='Buy & Hold', alpha=0.7)
ax.set_title('Comparaison des métriques')
ax.set_xlabel('Métrique')
ax.set_ylabel('Valeur')
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Sauvegarde des résultats
results_dir = '../results'
os.makedirs(results_dir, exist_ok=True)

results = {
    'symbol': symbol,
    'strategy_performance': performance,
    'buy_hold_performance': buy_hold_performance,
    'trades': [t for t in strategy.trades],
    'equity_curve': strategy.equity,
    'timestamp': datetime.now().isoformat(),
}

results_path = os.path.join(results_dir, f'{symbol}_backtest_results.pkl')
with open(results_path, 'wb') as f:
    pickle.dump(results, f)

print(f"Résultats sauvegardés: {results_path}")

# Sauvegarde en CSV
csv_path = os.path.join(results_dir, f'{symbol}_backtest_results.csv')
pd.DataFrame({
    'equity': strategy.equity,
    'drawdown': drawdown,
}).to_csv(csv_path)

print(f"CSV sauvegardé: {csv_path}")

print("\n" + "=" * 50)
print("Backtesting terminé!")
print("=" * 50)
```
