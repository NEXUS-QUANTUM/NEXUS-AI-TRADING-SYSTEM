```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("NEXUS AI TRADING SYSTEM - Feature Engineering")
print("=" * 50)
```

## 1. Chargement des Données

```python
# Chargement des données
import os
data_dir = '../data/processed'

data = {}
for file in os.listdir(data_dir):
    if file.endswith('_processed.csv'):
        symbol = file.replace('_processed.csv', '')
        df = pd.read_csv(os.path.join(data_dir, file), index_col=0, parse_dates=True)
        data[symbol] = df
        print(f"Chargé: {symbol} ({len(df)} observations)")
```

## 2. Features de Prix

```python
def add_price_features(df):
    """Ajoute les features basées sur les prix"""
    df = df.copy()
    
    # Rendements
    for period in [1, 5, 10, 20]:
        df[f'return_{period}d'] = df['close'].pct_change(period)
        df[f'log_return_{period}d'] = np.log(df['close'] / df['close'].shift(period))
    
    # Prix normalisés
    df['price_normalized'] = (df['close'] - df['close'].rolling(20).mean()) / df['close'].rolling(20).std()
    
    # Range
    df['high_low_range'] = (df['high'] - df['low']) / df['close']
    df['open_close_range'] = (df['close'] - df['open']) / df['open']
    
    # Features de momentum
    for period in [5, 10, 20]:
        df[f'momentum_{period}'] = df['close'] / df['close'].shift(period) - 1
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_price_features(df)

print("Features de prix ajoutées")
print(data['BTC-USD'].head())
```

## 3. Features de Volatilité

```python
def add_volatility_features(df):
    """Ajoute les features de volatilité"""
    df = df.copy()
    
    # Volatilité historique
    for window in [5, 10, 20, 50]:
        df[f'volatility_{window}d'] = df['close'].pct_change().rolling(window).std()
    
    # Volatilité normalisée
    df['volatility_ratio'] = df['volatility_10d'] / df['volatility_50d']
    
    # ATR (Average True Range)
    high_low = df['high'] - df['low']
    high_close = np.abs(df['high'] - df['close'].shift())
    low_close = np.abs(df['low'] - df['close'].shift())
    tr = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
    df['atr'] = tr.rolling(14).mean()
    df['atr_normalized'] = df['atr'] / df['close']
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_volatility_features(df)

print("Features de volatilité ajoutées")
print(data['BTC-USD'].head())
```

## 4. Features de Volume

```python
def add_volume_features(df):
    """Ajoute les features de volume"""
    df = df.copy()
    
    # Volume normalisé
    df['volume_ma_20'] = df['volume'].rolling(20).mean()
    df['volume_ratio'] = df['volume'] / df['volume_ma_20']
    
    # Volume price trend
    df['vpt'] = (df['volume'] * df['close'].pct_change()).cumsum()
    
    # On-balance volume
    df['obv'] = (np.sign(df['close'].diff()) * df['volume']).cumsum()
    
    # Money Flow Index
    typical_price = (df['high'] + df['low'] + df['close']) / 3
    money_flow = typical_price * df['volume']
    positive_flow = money_flow.where(df['typical_price'] > df['typical_price'].shift(), 0).rolling(14).sum()
    negative_flow = money_flow.where(df['typical_price'] < df['typical_price'].shift(), 0).rolling(14).sum()
    mfi = 100 - (100 / (1 + positive_flow / negative_flow))
    df['mfi'] = mfi
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_volume_features(df)

print("Features de volume ajoutées")
print(data['BTC-USD'].head())
```

## 5. Features Temporelles

```python
def add_temporal_features(df):
    """Ajoute les features temporelles"""
    df = df.copy()
    
    # Day of week
    df['day_of_week'] = df.index.dayofweek
    df['day_of_month'] = df.index.day
    df['month'] = df.index.month
    df['quarter'] = df.index.quarter
    df['year'] = df.index.year
    
    # Cyclical encoding
    df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_temporal_features(df)

print("Features temporelles ajoutées")
print(data['BTC-USD'].head())
```

## 6. Features de Lag

```python
def add_lag_features(df, target='close', lags=[1, 2, 3, 5, 10]):
    """Ajoute les features de lag"""
    df = df.copy()
    
    for lag in lags:
        df[f'{target}_lag_{lag}'] = df[target].shift(lag)
        df[f'{target}_lag_{lag}_diff'] = df[target] - df[target].shift(lag)
        df[f'{target}_lag_{lag}_pct'] = df[target] / df[target].shift(lag) - 1
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_lag_features(df)

print("Features de lag ajoutées")
print(data['BTC-USD'].head())
```

## 7. Features de Technical Indicators

```python
def add_technical_features(df):
    """Ajoute les features d'indicateurs techniques avancés"""
    df = df.copy()
    
    # Stochastic Oscillator
    low_14 = df['low'].rolling(14).min()
    high_14 = df['high'].rolling(14).max()
    df['stochastic_k'] = 100 * (df['close'] - low_14) / (high_14 - low_14)
    df['stochastic_d'] = df['stochastic_k'].rolling(3).mean()
    
    # Williams %R
    df['williams_r'] = -100 * (high_14 - df['close']) / (high_14 - low_14)
    
    # CCI
    typical_price = (df['high'] + df['low'] + df['close']) / 3
    sma = typical_price.rolling(20).mean()
    mad = typical_price.rolling(20).apply(lambda x: np.abs(x - x.mean()).mean())
    df['cci'] = (typical_price - sma) / (0.015 * mad)
    
    # ROC
    df['roc'] = (df['close'] / df['close'].shift(10) - 1) * 100
    
    # ADX
    tr = pd.concat([
        df['high'] - df['low'],
        np.abs(df['high'] - df['close'].shift()),
        np.abs(df['low'] - df['close'].shift())
    ], axis=1).max(axis=1)
    
    up_move = df['high'] - df['high'].shift()
    down_move = df['low'].shift() - df['low']
    
    plus_dm = np.where((up_move > down_move) & (up_move > 0), up_move, 0)
    minus_dm = np.where((down_move > up_move) & (down_move > 0), down_move, 0)
    
    atr = tr.rolling(14).mean()
    plus_di = 100 * (pd.Series(plus_dm).rolling(14).mean() / atr)
    minus_di = 100 * (pd.Series(minus_dm).rolling(14).mean() / atr)
    dx = 100 * np.abs(plus_di - minus_di) / (plus_di + minus_di)
    df['adx'] = dx.rolling(14).mean()
    
    return df

# Application
for symbol, df in data.items():
    data[symbol] = add_technical_features(df)

print("Features techniques avancées ajoutées")
print(data['BTC-USD'].head())
```

## 8. Normalisation des Features

```python
def normalize_features(df, method='standard'):
    """Normalise les features"""
    df = df.copy()
    
    # Sélection des features numériques
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    
    if method == 'standard':
        scaler = StandardScaler()
        df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    elif method == 'robust':
        scaler = RobustScaler()
        df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
    
    return df, scaler

# Normalisation
normalized_data = {}
scalers = {}

for symbol, df in data.items():
    df_norm, scaler = normalize_features(df, method='standard')
    normalized_data[symbol] = df_norm
    scalers[symbol] = scaler

print("Features normalisées")
print(normalized_data['BTC-USD'].head())
```

## 9. Analyse des Features

```python
# Analyse des corrélations des features
df = normalized_data['BTC-USD']
feature_cols = [col for col in df.columns if col not in ['open', 'high', 'low', 'close', 'volume']]

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Distribution des features
ax = axes[0]
sample_features = feature_cols[:10]
df[sample_features].boxplot(ax=ax)
ax.set_title('Distribution des features')
ax.set_xlabel('Features')
ax.set_ylabel('Valeur normalisée')
ax.xticks(rotation=45)

# Corrélation avec le prix
ax = axes[1]
correlations = df[feature_cols].corrwith(df['close']).sort_values()
correlations.tail(20).plot(kind='barh', ax=ax)
ax.set_title('Top 20 features corrélées avec le prix')
ax.set_xlabel('Corrélation')

plt.tight_layout()
plt.show()
```

```python
# Importance des features (PCA)
X = df[feature_cols].dropna()
pca = PCA(n_components=10)
pca_result = pca.fit_transform(X)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(range(1, 11), pca.explained_variance_ratio_.cumsum(), 'bo-')
plt.xlabel('Nombre de composantes')
plt.ylabel('Variance expliquée cumulée')
plt.title('Analyse PCA')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.bar(range(1, 11), pca.explained_variance_ratio_)
plt.xlabel('Composante')
plt.ylabel('Variance expliquée')
plt.title('Variance par composante')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Variance expliquée par 10 composantes: {pca.explained_variance_ratio_.cumsum()[-1]:.4f}")
```

## 10. Sauvegarde des Features

```python
# Sauvegarde des données
import pickle
import os

output_dir = '../data/features'
os.makedirs(output_dir, exist_ok=True)

for symbol, df in normalized_data.items():
    filepath = os.path.join(output_dir, f"{symbol}_features.csv")
    df.to_csv(filepath)
    print(f"Features sauvegardées: {filepath}")

# Sauvegarde des scalers
scaler_path = os.path.join(output_dir, 'scalers.pkl')
with open(scaler_path, 'wb') as f:
    pickle.dump(scalers, f)
print(f"Scalers sauvegardés: {scaler_path}")

# Sauvegarde des noms de features
feature_names = [col for col in df.columns if col not in ['open', 'high', 'low', 'close', 'volume']]
feature_path = os.path.join(output_dir, 'feature_names.txt')
with open(feature_path, 'w') as f:
    for name in feature_names:
        f.write(f"{name}\n")
print(f"Noms des features sauvegardés: {feature_path}")

print("\n" + "=" * 50)
print(f"Total features: {len(feature_names)}")
print("Feature engineering terminée!")
print("=" * 50)
```
